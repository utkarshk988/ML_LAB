```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#STEP 2
data = {
    'Customer ID' : list(range(1,21)),
    'Gender' : ['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'Age' : [19, 21, 20, 23, 31, 22, 35, 23, 64, 30, 67, 35, 58, 24, 37, 22, 35, 20, 52, 35],
    'Annual Income(k$)' : [15, 15, 16, 16, 17, 17, 18, 18, 19, 19, 19, 19, 20, 20, 20, 20, 21, 21, 23, 23],
    'Spending Score (1-100)' : [39, 81, 6, 77, 40, 76, 6, 94, 3, 72, 14, 99, 15, 77, 13, 79, 35, 66, 29, 98]
}
```

```python
df = pd.DataFrame(data)
```

```python
#STEP 3: Select features ( Annual Income & Spending Score)
X = df[['Annual Income(k$)','Spending Score (1-100)']]
```

```python
# Step 4: Elbow Method to find optimal K
wcss = []
for i in range (1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
