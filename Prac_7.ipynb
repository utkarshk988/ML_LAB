```python
# to explore and implement  naive bayes classifier
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
# Create the dataset
data ={

       'Cheap'    : ['Yes', 'Yes' ,'No', 'Yes', 'No' ],
       'Promo' : ['Yes', 'No', 'Yes', 'No', 'No'],
       'Buy?' : ['Yes', 'Yes', 'No', 'Yes', 'No']
}
```

```python
df = pd.DataFrame(data)
```

```python
#step 2
le_cheap = LabelEncoder()
le_promo = LabelEncoder()
le_buy = LabelEncoder()
```

```python
df['Cheap'] = le_cheap.fit_transform(df['Cheap'])
df['Promo'] = le_promo.fit_transform(df['Promo'])
df['Buy?'] = le_buy.fit_transform(df['Buy?'])
```

```python
#step 3 : feature (X) and target (y)
X = df[['Cheap', 'Promo']]
y = df[['Buy?']]
```

```python
# step 4 : Train Naive Bayes
model = CategoricalNB()
model.fit(X, y)
```

```python
# Step 5 : Predict on the same data
predictions = model.predict(X)
```

```python
#Step 6 : Show results with decoded labels
df['Predicted_Buy?'] = le_buy.inverse_transform(predictions)
df['Actual_Buy?'] = le_buy.inverse_transform(y)
```
